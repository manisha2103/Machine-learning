```python
import os
from PIL import Image
import numpy as np
import pandas as pd
```

# resizing the data 

```python
label = []
target = []
file = "/content/digits/New folder/"
os.listdir(file)
for files in os.listdir(file):
    for file_1 in os.listdir(os.path.join(file,files)):
        file_2 = os.path.join(os.path.join(file,files),file_1)
        img = Image.open(file_2)
        img_1 = img.resize((8,8))
        img_2 = np.array(img_1).ravel()
        label.append(img_2)
        target.append(files)
```

```python
df = pd.DataFrame(label, columns = ["pixel"+str(i) for i in range(0,64)])
df["label"] = target
```

```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/mnist.zip')
```

```python
df.head()
```

```python
df.drop("Unnamed: 0",axis=1,inplace=True)
```

```python
df.label.value_counts()
```

```python
import seaborn as sns
sns.countplot(data=df, x='label')
```

# segregating the input and output

```python
X = df.drop('label', axis=1)

y = df['label']
```

# splitting the dataset into train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

# logistic regression

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
y_test_pred = classifier.predict(X_test)
from sklearn import metrics

lg=metrics.accuracy_score(y_test, y_test_pred)
lg
```

# decision tree classifier

```python
from sklearn.tree import DecisionTreeClassifier

dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train, y_train)

y_test_pred = dt_classifier.predict(X_test)
from sklearn import metrics

dt=metrics.accuracy_score(y_test, y_test_pred)
dt
```

# KNN classifier

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier()
classifier.fit(X_train, y_train)

y_test_pred = classifier.predict(X_test)

knn=metrics.accuracy_score(y_test, y_test_pred)
knn
```

# random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(X_train, y_train)

y_test_pred = classifier.predict(X_test)

from sklearn import metrics

rt=metrics.accuracy_score(y_test, y_test_pred)
rt
```

# Gaussian NB classifier

```python
from sklearn.naive_bayes import GaussianNB
classifier=GaussianNB()
classifier.fit(X_train,y_train)
y_test_pred = classifier.predict(X_test)
from sklearn import metrics
nb=metrics.accuracy_score(y_test, y_test_pred)
nb
```

# ADA BOOST classifier

```python
from sklearn.ensemble import AdaBoostClassifier
ada = AdaBoostClassifier()
ada.fit(X_train, y_train)
y_test_pred = ada.predict(X_test)
ada = metrics.accuracy_score(y_test, y_test_pred)
ada
```

```python
rows=[['Logistic Regression',lg],['DT classifier',dt],['KNN classifier',knn],['GaussianNB',nb],['Random forest Classifier',rt],['ADA BOOST classifier',ada]]

df1=pd.DataFrame(data=rows,columns=['TYPE OF CLASSIFIER','Accuracy_Score'])

df1
```
